```python
# here's the code to run
# this code should work independently of the previous code. 

# these threshold will have to be changed
bacteria_threshold = 500
virus_threshold = 80 # percent genome found

import pandas as pd
import numpy as np
import csv as csv
import matplotlib.pyplot as plt
import os.path

%matplotlib inline

counts_together = {} # counts using names as tuples

bacteria_df = pd.read_csv('bacteria_virus_data_machine.csv', header=0, encoding='utf-8')
bacteria_df = bacteria_df.rename(columns={'\ufeffKingdom': 'Kingdom'})

for col in list(bacteria_df):
    # checking if there is an associated virus file
    virus_file_path = 'virus_files/' + col + '.csv'
    if os.path.exists(virus_file_path):
        virus_df = pd.read_csv(virus_file_path, header=0, skiprows=[1], sep='\t')
        # now select the viruses that meet the threshold
        virus_thresholded_df = virus_df[virus_df["Percent Proteins Hit"] >= virus_threshold]
        # and now the bacteria in the file above the threshold
        bacteria_thresholded_df = bacteria_df[bacteria_df[col] >= bacteria_threshold] 
        # get descriptive names/labels
        virus_thresholded_df['name'] = \
            virus_thresholded_df["Accession Number"] + ', ' + \
            virus_thresholded_df["Spp"]  
        bacteria_thresholded_df['name'] = \
            bacteria_thresholded_df['Kingdom'] + ',' + \
            bacteria_thresholded_df['Phylum'] + ',' + \
            bacteria_thresholded_df['Class'] + ',' + \
            bacteria_thresholded_df['Order'] + ',' + \
            bacteria_thresholded_df['Family'] + ',' + \
            bacteria_thresholded_df['Genus']
        # now add these names to the counts list
        for vn in virus_thresholded_df['name']:
            for bn in bacteria_thresholded_df['name']:
                name_tuple = (vn,bn)
                if name_tuple in counts_together:
                    counts_together[name_tuple] += 1
                else:
                    counts_together[name_tuple] = 1

# reordering counts_together and printing to a tab-delimited file                    
d = counts_together # shortening the name for a bit
sorted_counts_together = [(k, d[k]) for k in sorted(d, key=d.get, reverse=True)]

# now output that to a file
counts_file = 'cooccurance.tsv'
with open(counts_file, 'w') as f:
    for k,v in sorted_counts_together:
        display_string = str(v) + "\t" + k[0] + "\t" + k[1]
        print(display_string)
        f.write(display_string + "\n")
```
